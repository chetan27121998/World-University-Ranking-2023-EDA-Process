# **About Dataset**

The World University Rankings 2023 dataset include 1,799 universities across 104 countries and regions, making them the largest and most diverse university rankings to date. The table is based on 13 carefully calibrated performance indicators that measure an institution’s performance across four areas: teaching, research, knowledge transfer and international outlook. This year’s ranking analyzed over 121 million citations across more than 15.5 million research publications and included survey responses from 40,000 scholars globally. Overall, we collected over 680,000 datapoints from more than 2,500 institutions that submitted data

```python
def analyze_nulls(df):
    """Analyzes null values in a Pandas DataFrame and provides a descriptive summary.

    Args:
      df: The input Pandas DataFrame.

    Returns:
      A Pandas DataFrame summarizing the null values for each column, including:
        - Column Name: The name of the column.
        - Null Count: The number of null values in the column.
        - Null Percentage: The percentage of null values in the column.
        - Null Description: A textual description of the null values,
                           considering the column's data type and potential implications.
    """

    null_analysis = pd.DataFrame(columns=['Column Name', 'Null Count', 'Null Percentage', 'Null Description'])

    for col in df.columns:
        null_count = df[col].isnull().sum()
        null_percentage = (null_count / len(df)) * 100

        # Generate a more detailed description of the nulls
        description = f"{null_count} null values ({null_percentage:.2f}%). "
        if null_count > 0:
          if df[col].dtype in ['int64', 'float64']:
              description += "Missing numerical data, consider imputation or removal."
          elif df[col].dtype == 'object':  # Categorical
              description += "Missing categorical data, consider imputation with mode, a new category, or removal."
          else:
              description += "Missing data - further investigation of data type is needed to determine handling strategy."
        else:
            description += "No missing values in this column"


        null_analysis = pd.concat([null_analysis, pd.DataFrame({
            'Column Name': [col],
            'Null Count': [null_count],
            'Null Percentage': [null_percentage],
            'Null Description': [description]
        })], ignore_index=True)

    return null_analysis


# Example usage (assuming 'df' is your DataFrame):
null_summary = analyze_nulls(df)
null_summary
```

```python
# 1. Count of universities per location
universities_per_location = df['Location'].value_counts()
print("University Counts by Location:\n", universities_per_location)

# 2. Convert 'OverAll Score' to numeric
df['OverAll Score'] = pd.to_numeric(df['OverAll Score'], errors='coerce')

# 3. Compute the average overall score per location
average_scores = df.groupby('Location').agg({'OverAll Score': 'mean'})
print("\nAverage Scores by Location:\n", average_scores)

# 4. Identify the top 10 universities by overall score
top_universities = df.sort_values(by='OverAll Score', ascending=False).head(10)
print("\nTop 10 Universities by Overall Score:\n", top_universities)

# 5. Visualize distribution of 'OverAll Score'
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='OverAll Score', kde=True, bins=30)
plt.title('Overall Score Distribution')
plt.xlabel('Overall Score')
plt.ylabel('Number of Universities')
plt.grid(axis='y', linestyle='--')
plt.show()

# 6. Compute correlation matrix for numerical columns
numeric_data = df.select_dtypes(include=['number'])
correlation_data = numeric_data.corr()

# Visualize the correlation matrix as a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_data, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Between Numerical Features')
plt.show()

# 7. Box plots for scores in specific categories
categories = ['Teaching Score', 'Research Score', 'Citations Score',
              'Industry Income Score', 'International Outlook Score']

plt.figure(figsize=(16, 8))
for index, category in enumerate(categories, 1):
    plt.subplot(2, 3, index)
    sns.boxplot(y=df[category])
    plt.title(f'Distribution of {category}')
    plt.ylabel('')
plt.tight_layout()
plt.show()
```
